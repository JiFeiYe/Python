```python
# 员工离职检测
import pandas as pd
from sklearn.model_selection import train_test_split
import math

# 读取数据到pandas dataframe
df = pd.read_csv('HR_comma_sep.csv')

# 检测是否有缺失数据
df.isnull().any()
```

```python
# 查看数据样例
df.head()
```

```python
# 计算出离职员工的百分比和没有离职的百分比
left_rate = df['left'].value_counts() / len(df)

print("样本数据中，离职率为：%.2f\n\n" % left_rate[1])

# 显示统计数据
df.describe()
```

```python
# 将字符串转换为整数类型，不然后面无法计算
df['Department'] = df['Department'].astype('category').cat.codes
df['salary'] = df['salary'].astype('category').cat.codes

# 将数据分为训练集和测试集
X = df.drop(['left'], axis=1)
y = df['left']

# 注意参数stratify=y，意味着在产生训练数据和测试数据中，离职员工的百分比等于原来总数据中离职员工的百分比
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=123, stratify=y)
```

```python
# 训练模型 决策树
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree

# 训练决策树
dtree = DecisionTreeClassifier(
    # TODO 1：编写代码实现各种参数的选择和配置 
    criterion='gini',
    max_depth=None,
    min_samples_leaf=5,
    min_samples_split=5,
    random_state=42
)
dtree.fit(X_train, y_train)
print("---决策树---")
print(classification_report(y_test, dtree.predict(X_test)))
```
